```python
import numpy
import re
import os
```

```python
def write_toki(file_str):

    h = 0
    n = 0
    dens = []
    temp = []

    elements = re.findall('([0-9]{1,2})([A-Za-z]{1,2})', file_str)
    el1 = elements[0][1] + elements[0][0]
    el2 = elements[1][1] + elements[1][0]

    id =f"       {el1} {el2}"

    toki_file = f"{el1}--{el2}-toki"

    with open(file_str, 'r') as file:
        for line in file:
            if line.startswith('!'):
                h +=1
            else:
                numb = line.strip().split()
                dens.append(float(numb[0]))
                temp.append(float(numb[1]))
                n +=1
                            
    den = sorted(set(dens))
    tem = sorted(set(temp))

    den_points = len(den)
    tem_points = len(tem)

    with open(toki_file, 'w') as output:
        output.write("t\n")
        output.write(f"{id}\n")
        output.write(f"{file_str}\n")
        output.write(f"{h}\n")
        output.write(f"{den_points}\n")
        output.write(f"{tem_points}\n")
```

```python
for _,_,files in os.walk(os.getcwd()):
    for file in files:
        if re.search('[\w\d]*_[\w]*\.dat', file):
            print(file)
            write_toki(file)
```
