# Example of multiple output channels of a reaction

## Here we look at O16 + O16

```python
import pynucastro as pyna
```

```python
mylibrary = pyna.ReacLibLibrary()
```

```python
o16_exact_filter = pyna.rates.RateFilter(reactants=['o16', 'o16'])
```

```python
o16_rates = mylibrary.filter(o16_exact_filter)
```

```python
o16_rates
```

```python
o16o16a = o16_rates.get_rate_by_name("o16(o16,n)s31")
o16o16b = o16_rates.get_rate_by_name("o16(o16,p)p31")
o16o16c = o16_rates.get_rate_by_name("o16(o16,a)si28")
```

```python
print(o16o16a.original_source)
```

```python
print(o16o16b.original_source)
```

```python
print(o16o16c.original_source)
```

```python
fig = o16o16a.plot()
```

```python
fig = o16o16b.plot()
```

```python
fig = o16o16c.plot()
```
